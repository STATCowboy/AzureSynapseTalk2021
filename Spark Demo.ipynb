```python
%%pyspark
from azureml.opendatasets import NycTlcYellow
from dateutil import parser

end_date = parser.parse('2018-06-28')
start_date = parser.parse('2018-06-21')
nyc_tlc = NycTlcYellow(start_date=start_date, end_date=end_date)
nyc_tlc_df = nyc_tlc.to_spark_dataframe()
```

```python
%%pyspark
nyc_tlc_df.printSchema()
```

```python
%%pyspark
spark.sql("CREATE DATABASE IF NOT EXISTS nyctaxi")
nyc_tlc_df.write.mode("overwrite").saveAsTable("nyctaxi.trip")
```

```python
%%pyspark
nyc_tlc_df = spark.sql("SELECT * FROM nyctaxi.trip") 
display(nyc_tlc_df)
```

```python
%%pyspark
nyc_tlc_df = spark.sql("""
   SELECT PassengerCount,
       SUM(tripDistance) as SumTripDistance,
       AVG(tripDistance) as AvgTripDistance
   FROM nyctaxi.trip
   WHERE tripDistance > 0 AND PassengerCount > 0
   GROUP BY PassengerCount
   ORDER BY PassengerCount
""") 
display(nyc_tlc_df)
nyc_tlc_df.write.mode("overwrite").saveAsTable("nyctaxi.passengercountstats")
```
